```
movies_df = pd.read_csv('movies.csv')
ratings_df = pd.read_csv('IMDb_ratings.csv')
movies_df = movies_df.merge(ratings_df, on='imdb_title_id', how='inner')
movies_df.head()
```

```
movies_df = movies_df[movies_df.date_published.notnull()]
sns.set(style="white")
plt.figure(figsize=(15,10))
plt.title("Movies v/s Countries")
count = movies_df['country'].value_counts()[:10]
sns.barplot(count.index, count.values)
plt.xlabel('Country')
plt.ylabel('Number of movies')
```

```
movies_df = movies_df[movies_df.date_published.notnull()]
sns.set(style="dark")
plt.figure(figsize=(25,10))
plt.title("Movies v/s Language")
count = movies_df['language'].value_counts()[:9]
sns.barplot(count.index, count.values)
plt.xlabel('Language')
plt.ylabel('Number of movies')
```

```
k = 100
plt.figure(figsize=(25,10))
count = movies_df['production_company'].value_counts()[:30]
sns.barplot(count.values, count.index)
for i, v in enumerate(count.values):
    plt.text(0.8,i,v,color='k')
plt.xlabel('Count', fontsize=12)
plt.ylabel('Production Company name')
plt.title("Distribution of Studio names")
```

```
movies_df['first_genre'] = movies_df['genre'].str.split(',').str[0]

plt.figure(figsize=(25,15))
count = movies_df['first_genre'].value_counts()[:20]
sns.barplot(count.values, count.index)
for i, v in enumerate(count.values):
    plt.text(0.8,i,v)
plt.xlabel('Count')
plt.ylabel('Genre name')
plt.title("Distribution of Genres")
```

```
top_genres = list(count.index)
movie_genres_df = movies_df[movies_df['first_genre'].isin(top_genres)]
movie_genres_df = movie_genres_df[pd.notnull(movie_genres_df[['first_genre', 'avg_vote', 'votes']])]

plt.figure(figsize=(25, 10))
sns.boxplot(x='first_genre', y='avg_vote', data=movie_genres_df)
plt.xlabel("Genre Name",fontsize=14)
plt.ylabel("IMDb Average Rating",fontsize=14)
plt.title("Boxplot of average rating per Genre", fontsize=16)
plt.show()
```

```
movies_df.head(10)
indian_mov_df=movies_df[movies_df["country"]=="India"]
indian_mov_df.head()
```

```
movies_df['first_genre'] = movies_df['genre'].str.split(',').str[0]
count = movies_df['first_genre'].value_counts()[:20]
top_genres = list(count.index)
indian_mov_genres_df = indian_mov_df[indian_mov_df['first_genre'].isin(top_genres)]
indian_mov_genres_df = indian_mov_genres_df[pd.notnull(indian_mov_genres_df[['first_genre', 'avg_vote', 'votes']])]

plt.figure(figsize=(25, 10))
sns.boxplot(x='first_genre', y='avg_vote', data=indian_mov_genres_df)
plt.xlabel("Genre Name",fontsize=14)
plt.ylabel("IMDb Average Rating",fontsize=14)
plt.title("Movie Genres that indians love", fontsize=16)
plt.show()
```
